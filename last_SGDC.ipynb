```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('new_clean.csv',low_memory=False)
```

```python
df['Label'].value_counts()
```

```python
x=df.iloc[:,df.columns != 'Label']
y=df.iloc[:,-1]
print("x\n",x.info())
y = pd.DataFrame(y)
print('y\n',y.info())
```

```python
normalized_x=(x-x.mean())/x.std()
normalized_x = normalized_x.drop('Unnamed: 0',axis=1)
```

```python
normalized_x.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(normalized_x, y, test_size=0.25, random_state=0)
y_train = np.array(y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=8)
model.fit(X_train, y_train)
```
